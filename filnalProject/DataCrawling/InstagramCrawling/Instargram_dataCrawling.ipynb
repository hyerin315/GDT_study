```python
driver.get(f'https://www.instagram.com/explore/tags/{hashTag}/')
```

```python
# 데이터를 저장할 Dictionary
insta_dict = {'id':[],
              'date': [], # datetime 형식
              #'date_title': [], # time > title 형식
              'like': [],
              'text': [],
              'hashtag':[]}
 
 
# 첫 번째 게시물 클릭 
first_post = driver.find_element_by_class_name('eLAPa')
first_post.click()
time.sleep(2.5)

i = 1
while i < 1339 :
    driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow').click()
    time.sleep(1.5)
    i += 1
```

```python
seq = 0
start = time.time()
contents = 1339

while True:
    try:
        if driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow'):
            
            # 시간정보 수집 
            time_raw = driver.find_element_by_css_selector('time.FH9sR.Nzb55')
            time_edit1 = (time_raw.get_attribute('datetime')) # append용
            time_edit2 = (time_raw.get_attribute('datetime'))[:10] # 비교용
            
            time_info1 = pd.to_datetime(time_edit1)
            print(time_info1)
            time_info2 = pd.to_datetime(time_edit2) # 타임스탬프
            time_please = time_info2.to_pydatetime()
#             print(type(time_please))
#             print(time_please)
            
            time01 = datetime.datetime(2021, 7, 15)
            time02 = datetime.datetime(2021, 8, 15)
#             print(time01)
            
            if (time_please >= time01) and (time_please <= time02) :
#                 print("성공")
            
                if seq % 20 == 0:
                    print('{}번째 수집 중'.format(seq), time.time() - start, sep = '\t')


                # id 정보 수집
                try:
                    info_id = driver.find_element_by_css_selector('h2._6lAjh').text
                    insta_dict['id'].append(info_id)
                except:
                    info_id = driver.find_element_by_css_selector('div.C4VMK').text.split()[0]
                    insta_dict['id'].append(info_id)

                # 시간정보 수집 
                insta_dict['date'].append(time_info1)
                
                    
                # like 정보 수집
                try:
                    like = driver.find_element_by_css_selector('a.zV_Nj > span').text
                    insta_dict['like'].append(like)

                except:
                    insta_dict['like'].append('영상')



                # text 정보수집
                raw_info = driver.find_element_by_css_selector('div.C4VMK').text.split()
                text = []
                for i in range(len(raw_info)):
                    ## 첫번째 text는 아이디니까 제외 
                    if i == 0:
                        pass
                    ## 두번째부터 시작 
                    else:
                        if '#' in raw_info[i]:
                            pass
                        else:
                            text.append(raw_info[i])
                            
                clean_text = ' '.join(text)
                insta_dict['text'].append(clean_text)

                # hashtag 수집
                raw_tags = driver.find_elements_by_css_selector('a.xil3i')
                hash_tag = []
                for i in range(len(raw_tags)):
                    if raw_tags[i].text == '':
                        pass
                    else:
                        hash_tag.append(raw_tags[i].text)

                insta_dict['hashtag'].append(hash_tag)

                driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow').click()
                time.sleep(3)

                seq += 1
                contents += 1
                if contents % 20 == 0:
                    print('{}번째 콘텐츠'.format(contents))

#                 if seq == 5:
#                     break
                
            else:
#                 print("실패")
                contents += 1
                print('{}번째 콘텐츠'.format(contents), ' 에러1')

                driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow').click()
                time.sleep(3)
                
 
        else:
            contents += 1
            print('{}번째 콘텐츠'.format(contents), ' 에러2')
            break
            
    except:
        contents += 1
        print('{}번째 콘텐츠'.format(contents), end=" ")
        print("에러3")
        driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow').click()
        time.sleep(3.5)
```
